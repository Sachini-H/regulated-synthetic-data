```python
# Function to read a given file from a given folder in CSV format
def read_events(folder_path,file_path,start):
  csv_file_path = os.path.join(folder_path, file_path)
  df = pd.read_csv(csv_file_path)

  # For gid generation
  df['number'] = range(start, start+ len(df))

  # Convert column to datetime type
  df['evt_gmt'] = pd.to_datetime(df['evt_gmt'])

  # Convert datetime column to timestamp
  df['timestamp'] = df['evt_gmt'].apply(lambda x: x.timestamp())
  df['timestamp'] = df['timestamp'].astype(int)

  if 'buyer' in df.columns:
    df['gid_new']= df.apply(lambda row: f"device456@-{row['buyer']}-123-{row['style']}-@-" + f"{row['timestamp']}-{row['number']}", axis=1)
  return df

```
